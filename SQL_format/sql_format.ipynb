```python
from pathlib import Path
import re

# Leer un archivo de texto usando Pathlib
file_path = Path('test_sql_3.txt')
text = file_path.read_text()

print(type(text))
```

```python
# Transforma a mayusculas las palabras reservadas

def upper(sql):
    keywords = r'\b(SELECT|FROM|WHERE|WHEN|INSERT|INTO|VALUES|UPDATE|SET|DELETE|JOIN|INNER|LEFT|RIGHT|FULL|ON|END|AS|CREATE|CASE|TABLE|THEN|PRIMARY|KEY|FOREIGN|REFERENCES|ALTER|DROP|INDEX|VIEW|DISTINCT|GROUP BY|ORDER BY|HAVING|LIMIT|OFFSET|UNION|ALL|EXISTS|BETWEEN|LIKE|IN|NOT IN|IS|NULL|AND|OR|NOT|IFNULL|CAST)\b'
    
    def capitalize_keywords(match):
        return match.group(0).upper()

    formatted_sql = re.sub(keywords, capitalize_keywords, sql, flags=re.IGNORECASE)
    return formatted_sql

# Testing
formatted_query = upper(text)
print(formatted_query)
```

```python

import re

def insertar_linea(texto, nueva_linea_1, nueva_linea_2, nueva_linea_3):
    # Expresión regular para encontrar frases que comienzan con "with" y terminan con "as"
    patron = r'(\bwith\b.*?\bas\b)'
    # Insertar la nueva línea sobre las coincidencias
    texto_modificado = re.sub(patron, lambda m: nueva_linea_1 + "\n" + nueva_linea_2 + "\n" + nueva_linea_3 + "\n" + m.group(0), texto, flags=re.IGNORECASE | re.DOTALL)
    return texto_modificado

# Testing
nueva_linea_1 = "Query Target:"
nueva_linea_2 = "Query Product:"
nueva_linea_3 = "Query Soruces:"
nueva_linea_4 = "#________________________________________________________________:"
texto_modificado = insertar_linea(text, nueva_linea_1, nueva_linea_2, nueva_linea_3)
print(texto_modificado)

```
