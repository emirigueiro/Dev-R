```python
from binance.client import Client
import requests
import re
```

```python
# Define tu API Key y API Secret
api_key = 'TU_API_KEY'
api_secret = 'TU_API_SECRET'

def binance_price():
    # Inicializa el cliente de Binance
    client = Client(api_key, api_secret)
    
    # Obtiene las cotizaciones actuales de todas las criptomonedas
    prices = client.get_all_tickers()
    
    # Almacena los resultados en una lista
    results = []
    for price in prices:
        results.append(f"Symbol: {price['symbol']}, Price: {price['price']}")
    
    return results
```

```python
import requests
import re

# Define la URL base de la API de OKX
base_url = 'https://www.okx.com/api/v5/market/tickers'

def okx_price():
    # Define el parámetro del tipo de mercado (SPOT en este caso)
    params = {
        'instType': 'SPOT'
    }
    
    # Realiza la solicitud GET a la API de OKX
    response = requests.get(base_url, params=params)
    
    # Verifica si la solicitud fue exitosa
    if response.status_code == 200:
        data = response.json()
        tickers = data['data']
        # Lista para almacenar los símbolos procesados
        symbols = []
        # Imprime las cotizaciones originales y extrae los símbolos
        for ticker in tickers:
            symbol = ticker['instId']
            price = ticker['last']
            symbols.append(f"Symbol: {symbol}, Price: {price}")           
    else:
        print(f"Error: Unable to fetch data (Status code: {response.status_code})")
        return []

    # Función para eliminar subcadenas entre '-' y ','
    def remove_substrings(lst):
        # Define una expresión regular para encontrar el patrón entre '-' y ','
        pattern = re.compile(r'-')
        
        # Procesa cada elemento de la lista
        new_lst = []
        new_lst = [pattern.sub('', item) for item in lst]
        
        return new_lst

    # Aplica la función a la lista de símbolos
    cleaned_symbols = remove_substrings(symbols)
    
    # Imprime los símbolos procesados
    cleaned_symnols_final = []
    for symbol in cleaned_symbols:
        cleaned_symnols_final.append(symbol)
    
    return cleaned_symnols_final
```
