```python
# Inicializa un diccionario para el conteo
conteo_palabras = {}

# Utiliza un bucle for para recorrer la lista
for palabra in letra_lista:
    # Si la palabra ya está en el diccionario, incrementa el conteo
    if palabra in conteo_palabras:
        conteo_palabras[palabra] += 1
    # Si la palabra no está en el diccionario, agrégala con un conteo inicial de 1
    else:
        conteo_palabras[palabra] = 1

conteo_palabras_ordenado = dict(sorted(conteo_palabras.items(), key=lambda item: item[1], reverse=True))

# Imprime el resultado
for palabra, conteo in conteo_palabras_ordenado.items():
    print(f"{palabra}: {conteo} veces")
```
